
# Exploratory Data Analysis (EDA)
## Dataset Kendaraan
Notebook ini berisi proses EDA pada dataset spesifikasi kendaraan. Data ini mencakup informasi seperti merk, model, tahun, jenis bahan bakar, tenaga mesin, jenis transmisi, dan harga. Tujuan dari analisis ini adalah untuk memahami struktur data, memvisualisasikan distribusi variabel, dan menemukan hubungan antar variabel yang mungkin signifikan.

### Langkah Analisis
Analisis EDA dilakukan melalui beberapa langkah berikut:
1. Mengecek informasi umum tentang dataset dan memeriksa missing values.
2. Melakukan visualisasi distribusi variabel numerik.
3. Meninjau hubungan antara beberapa variabel utama seperti `Engine HP`, `Popularity`, dan `MSRP`.



# Import Dataset


```python
import pandas as pd

# Melakukan import dataset terlebih dahulu
file_path = '/Users/narasi29/Downloads/data-2.csv'
data = pd.read_csv(file_path)
```


## 1. Informasi Umum Dataset

Di sini kita akan mengecek dimensi dataset, tipe data, dan informasi nilai yang hilang untuk mengetahui struktur data.


```python

# Menampilkan info umum dari dataset
data.info()
```


## 2. Mengecek Missing Values dan Statistik Dasar
Setelah melihat info umum, kita akan mengecek jumlah missing values per kolom dan statistik dasar untuk variabel numerik.


```python

# Mengecek missing values dan statistik deskriptif
missing_values = data.isnull().sum()
data.describe()
missing_values
```


## 3. Visualisasi Distribusi Kolom Numerik Utama
Untuk lebih memahami distribusi variabel seperti `Engine HP`, `Popularity`, dan `MSRP`, kita akan membuat histogram dari masing-masing variabel.


```python

import matplotlib.pyplot as plt
import seaborn as sns

# Menampilkan histogram dari variabel utama
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(data['Engine HP'].dropna(), ax=axes[0], kde=True).set_title("Distribusi Engine HP")
sns.histplot(data['Popularity'].dropna(), ax=axes[1], kde=True).set_title("Distribusi Popularity")
sns.histplot(data['MSRP'].dropna(), ax=axes[2], kde=True).set_title("Distribusi MSRP")
plt.tight_layout()
plt.show()
```


## 4. Analisis Korelasi
Mengecek hubungan antara variabel numerik seperti `Engine HP`, `Popularity`, dan `MSRP` untuk melihat korelasi potensial di antara mereka.


```python

# Menampilkan heatmap korelasi untuk variabel numerik
correlation_matrix = data[['Engine HP', 'Popularity', 'MSRP']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm').set_title("Korelasi antara variabel")
plt.show()
```
