```python
import pandas as pd #type: ignore
import matplotlib.pyplot as plt #type: ignore
import seaborn as sns #type: ignore

file_path = '/Users/narasi29/Documents/Sem 7/Machine Learning/Week 4/Dataset/StudentsPerformance.csv'
data = pd.read_csv(file_path)

data.columns = ['gender', 'race_ethnicity', 'parental_education', 'lunch', 
                'test_preparation', 'math_score', 'reading_score', 'writing_score']

```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.histplot(data['math_score'], kde=True, color='blue', label='Math Score')
sns.histplot(data['reading_score'], kde=True, color='green', label='Reading Score')
sns.histplot(data['writing_score'], kde=True, color='red', label='Writing Score')
plt.title('Distribution of Scores in Math, Reading, and Writing')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
race_avg_scores = data.groupby('race_ethnicity')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
race_avg_scores = pd.melt(race_avg_scores, id_vars='race_ethnicity', value_vars=['math_score', 'reading_score', 'writing_score'], 
                          var_name='Subject', value_name='Average Score')

sns.barplot(x='race_ethnicity', y='Average Score', hue='Subject', data=race_avg_scores)
plt.title('Average Scores by Race/Ethnicity Group')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')
plt.legend(title='Subject')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
parental_avg_scores = data.groupby('parental_education')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
parental_avg_scores = pd.melt(parental_avg_scores, id_vars='parental_education', 
                              value_vars=['math_score', 'reading_score', 'writing_score'], 
                              var_name='Subject', value_name='Average Score')

sns.barplot(x='parental_education', y='Average Score', hue='Subject', data=parental_avg_scores)
plt.title('Average Scores by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.xticks(rotation=45)
plt.ylabel('Average Score')
plt.legend(title='Subject')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
lunch_avg_scores = data.groupby('lunch')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
lunch_avg_scores = pd.melt(lunch_avg_scores, id_vars='lunch', 
                           value_vars=['math_score', 'reading_score', 'writing_score'], 
                           var_name='Subject', value_name='Average Score')

sns.barplot(x='lunch', y='Average Score', hue='Subject', data=lunch_avg_scores)
plt.title('Average Scores by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Average Score')
plt.legend(title='Subject')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
prep_avg_scores = data.groupby('test_preparation')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
prep_avg_scores = pd.melt(prep_avg_scores, id_vars='test_preparation', 
                          value_vars=['math_score', 'reading_score', 'writing_score'], 
                          var_name='Subject', value_name='Average Score')

sns.barplot(x='test_preparation', y='Average Score', hue='Subject', data=prep_avg_scores)
plt.title('Average Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Average Score')
plt.legend(title='Subject')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
gender_avg_scores = data.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
gender_avg_scores = pd.melt(gender_avg_scores, id_vars='gender', 
                            value_vars=['math_score', 'reading_score', 'writing_score'], 
                            var_name='Subject', value_name='Average Score')

sns.barplot(x='gender', y='Average Score', hue='Subject', data=gender_avg_scores)
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.legend(title='Subject')
plt.show()
```

```python
plt.figure(figsize=(8, 8))

# Gender distribution
gender_counts = data['gender'].value_counts()
plt.subplot(2, 2, 1)
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')

# Race/Ethnicity distribution
race_counts = data['race_ethnicity'].value_counts()
plt.subplot(2, 2, 2)
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Race/Ethnicity Distribution')

# Lunch type distribution
lunch_counts = data['lunch'].value_counts()
plt.subplot(2, 2, 3)
plt.pie(lunch_counts, labels=lunch_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Lunch Type Distribution')

# Test preparation course distribution
prep_counts = data['test_preparation'].value_counts()
plt.subplot(2, 2, 4)
plt.pie(prep_counts, labels=prep_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Test Preparation Course Distribution')

plt.tight_layout()
plt.show()
```

## Berikut adalah identifikasi singkat dari setiap hasil visualisasi pada kode di atas:

Gender Distribution:
Visualisasi ini menunjukkan proporsi jenis kelamin dari dataset. Hasilnya menampilkan persentase jumlah siswa perempuan dan laki-laki, memberikan gambaran umum mengenai distribusi gender dalam populasi data.
Race/Ethnicity Distribution:
Pie chart ini menggambarkan proporsi siswa berdasarkan kelompok ras/etnis. Dengan visualisasi ini, kita bisa melihat seberapa banyak siswa yang termasuk dalam masing-masing kelompok, membantu memahami komposisi etnis dalam dataset.
Lunch Type Distribution:
Visualisasi ini menunjukkan distribusi siswa berdasarkan jenis makan siang yang mereka terima (standar atau gratis/diskon). Hasil ini penting untuk melihat hubungan potensial antara jenis makan siang dan performa akademik, jika ada.
Test Preparation Course Distribution:
Pie chart ini menunjukkan proporsi siswa yang telah menyelesaikan kursus persiapan ujian dibandingkan dengan mereka yang tidak. Visualisasi ini membantu mengidentifikasi apakah ada korelasi antara mengikuti kursus persiapan dengan hasil akademik siswa.
